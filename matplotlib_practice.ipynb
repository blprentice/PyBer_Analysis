```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 
'Oct', 'Nov', 'Dec']

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90,
20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title('PyBer Fare by Month')

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker='*', color='blue', linewidth=2, label='Boston')
# Create labels for the x and y axes
plt.xlabel('date')
plt.ylabel('Fare($)')
# Set the y liunit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title('PyBer Fare by Month')
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Create the plot
ax = plt.axes()
ax.plot(x_axis, y_axis, marker='D', color='green', linewidth=2, label='Boston')
# Create labels for the x and y axes
ax.set_xlabel('date')
ax.set_ylabel('Fare($)')
# Set the y liunit between 0 and 45
ax.set_ylim(0, 45)
# Create a title
ax.set_title('PyBer Fare by Month')
# Add a grid
ax.grid()
# Add the legend
ax.legend()
```

```python
# Making the bubbles larger by multiple of 3
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# Plot
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Python list comprehension method
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Using object-oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create bubble chart with object-oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Same as before, but fancier
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c='skyblue', s = [i * 3 for i in y_axis], alpha=0.8, edgecolors='black', linewidths=2, label='Boston')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlim(0, 50)
ax.invert_yaxis()
```

```python
# Baking a pie chart, har har
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Make it fancy
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,labels=x_axis, autopct='%.1f%%', shadow=True, startangle=90, explode=explode_values, counterclock=False, colors=colors)
plt.show()
```

```python
%matplotlib inline
# Import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
# Get the standard deviation of the values in the y-axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Object-oriented approach
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Adding error bars and caps to a bar chart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Same as above, but with object-oriented approach
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Adjust major x-axis ticks
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Object-oriented approach
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
# Import MultipleLocator method from matplotlib.ticker function
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
